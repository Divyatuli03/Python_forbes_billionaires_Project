```python
table = pd.pivot_table(data=df1,index='country',values='finalWorth',aggfunc=np.sum)
print(table)
```

```python
sorted_table=table.sort_values(by='finalWorth', ascending = False)
top_10_countries= sorted_table.head(10)
plt.bar(top_10_countries.index,top_10_countries['finalWorth'])
plt.xticks(rotation=70) 
plt.xlabel('Country') 
plt.ylabel('Final Worth') 
plt.title('Final Worth of top 10 countries') 
plt.savefig('country_wise_finalworth.jpg',dpi=300,bbox_inches='tight') 
plt.show()
```
